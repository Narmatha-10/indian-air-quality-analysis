### Author

# Project Overview

This project analyzes Indian Air Quality data collected from various monitoring stations across India.

The goal of the project is to:

- Clean and preprocess the data
- Explore pollution patterns
- Perform statistical analysis
- Create visualizations
- Build an interactive Streamlit dashboard

# Dataset Information

Source:
- Kaggle
- data.gov.in

Dataset contains:

- State
- City
- Station
- Pollutant Type
- Pollution Values
- Date and Time
- Latitude
- Longitude

Total Records: 4571

Total Columns: 11

# Data Cleaning Summary

- Converted date columns

# Exploratory Data Analysis

Key Findings:

- PM2.5 was one of the most frequently monitored pollutants.
- Some states recorded significantly higher pollution levels.
- Moderate and Poor categories were common.
- Pollution varied across cities and pollutants.
