```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
df = pd.read_csv("../data/cleaned_air_quality.csv")
# BASIC OVERVIEW
print("\nDataset Shape:")
print(df.shape)

print("\nColumn Names:")
print(df.columns.tolist())

print("\nFirst 5 Rows:")
print(df.head())

print("\nDataset Information:")
print(df.info())


```

```python
# UNIVARIATE ANALYSIS
print("UNIVARIATE ANALYSIS")

# Pollutant Distribution
print("\nPollutant Frequency:")
print(df['pollutant_id'].value_counts())

# Pollution Category Distribution
print("\nPollution Category Distribution:")
print(df['pollution_category'].value_counts())

# State Distribution
print("\nTop States by Records:")
print(df['state'].value_counts().head(10))

# Summary Statistics
print("\nPollution Statistics:")
print(df['pollutant_avg'].describe())
```

```python
# QUESTION 1
# Which Pollutant Appears Most Frequently?
print("\nQUESTION 1")
print("Which pollutant appears most frequently?")

q1 = df['pollutant_id'].value_counts()

print(q1)
```

```python
# QUESTION 2
# Which State Has Highest Average Pollution?
print("\nQUESTION 2")
print("Top 10 States by Average Pollution")

q2 = (
    df.groupby('state')['pollutant_avg']
    .mean()
    .sort_values(ascending=False)
    .head(10)
)

print(q2)
```

```python
# QUESTION 3
# Most Polluted Cities
print("\nQUESTION 3")
print("Top 10 Most Polluted Cities")

q3 = (
    df.groupby('city')['pollutant_avg']
    .mean()
    .sort_values(ascending=False)
    .head(10)
)

print(q3)
```

```python
# QUESTION 4
# Which Pollutant Has Highest Average Concentration?
print("\nQUESTION 4")

q4 = (
    df.groupby('pollutant_id')['pollutant_avg']
    .mean()
    .sort_values(ascending=False)
)

print(q4)
```

```python
# ==========================================================
# QUESTION 5
# Which State Has Most Monitoring Stations?
print("\nQUESTION 5")

q5 = (
    df.groupby('state')['station']
    .nunique()
    .sort_values(ascending=False)
    .head(10)
)

print(q5)
```

```python
# QUESTION 6
# Distribution of Pollution Categories
print("\nQUESTION 6")

q6 = df['pollution_category'].value_counts()

print(q6)
```

```python
# QUESTION 7
# Which Pollutant Has Highest Variability?
print("\nQUESTION 7")

q7 = (
    df.groupby('pollutant_id')['pollution_range']
    .mean()
    .sort_values(ascending=False)
)

print(q7)
```

```python
# QUESTION 9
# Pollutant Counts By City
print("\nQUESTION 9")

q9 = pd.crosstab(
    df['city'],
    df['pollutant_id']
)

print(q9.head())
```

```python
# QUESTION 10
# Peak Pollution Hour
print("\nQUESTION 10")

q10 = (
    df.groupby('update_hour')['pollutant_avg']
    .mean()
    .sort_values(ascending=False)
)

print(q10)
```

```python
# BIVARIATE ANALYSIS
print("BIVARIATE ANALYSIS")
# State vs Average Pollution
state_pollution = (
    df.groupby('state')['pollutant_avg']
    .mean()
    .sort_values(ascending=False)
)

print(state_pollution.head(10))

# Pollutant vs Average Pollution
pollutant_analysis = (
    df.groupby('pollutant_id')['pollutant_avg']
    .mean()
)

print(pollutant_analysis)

# Pollution Category vs Pollutant
category_pollutant = pd.crosstab(
    df['pollution_category'],
    df['pollutant_id']
)

print(category_pollutant)
```

```python
# MULTIVARIATE ANALYSIS
print("MULTIVARIATE ANALYSIS")

multi = pd.pivot_table(
    df,
    values='pollutant_avg',
    index='state',
    columns='pollutant_id',
    aggfunc='mean'
)

print(multi.head())
```

```python
# CORRELATION MATRIX
print("\nCorrelation Matrix")
corr = df[
    [
        'pollutant_min',
        'pollutant_max',
        'pollutant_avg',
        'pollution_range'
    ]
].corr()

print(corr)
```
