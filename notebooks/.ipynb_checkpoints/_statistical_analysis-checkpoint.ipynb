```python
# Statistical Analysis Summary

# Hypothesis Test 1
#Compared PM2.5 and PM10 pollution levels using an Independent T-Test.

# Hypothesis Test 2
#Compared pollution levels between North India and South India.

# Hypothesis Test 3
#Used a Chi-Square Test to determine whether pollution category depends on pollutant type.

## Correlation Analysis
#Analyzed relationships among:
#- Pollutant Minimum
#- Pollutant Maximum
#- Pollutant Average
#- Pollution Range
```

```python
# Import Libraries
import pandas as pd
import numpy as np
from scipy.stats import ttest_ind
from scipy.stats import chi2_contingency
from scipy.stats import pearsonr
import warnings
warnings.filterwarnings('ignore')

# LOAD DATASET
df = pd.read_csv("../data/cleaned_air_quality.csv")
print("DATASET LOADED SUCCESSFULLY")
print("\nDataset Shape:")
print(df.shape)
```

```python
# HYPOTHESIS TEST 1
# PM2.5 vs PM10
# Independent T-Test

print("HYPOTHESIS TEST 1")
print("PM2.5 vs PM10 Pollution Levels")

pm25 = df[df['pollutant_id'] == 'PM2.5']['pollutant_avg']

pm10 = df[df['pollutant_id'] == 'PM10']['pollutant_avg']

t_stat, p_value = ttest_ind(pm25, pm10)

print("T-Statistic:", t_stat)
print("P-Value:", p_value)

if p_value < 0.05:
    print("\nConclusion:")
    print("There is a statistically significant difference between PM2.5 and PM10 pollution levels.")
else:
    print("\nConclusion:")
    print("There is no statistically significant difference between PM2.5 and PM10 pollution levels.")
```

```python
# HYPOTHESIS TEST 2
# North India vs South India
# Independent T-Test
print("HYPOTHESIS TEST 2")
print("North India vs South India Pollution")
north_states = [
    'Delhi',
    'Haryana',
    'Punjab',
    'Uttar Pradesh'
]

south_states = [
    'Tamil Nadu',
    'Kerala',
    'Karnataka',
    'Andhra Pradesh'
]

north = df[
    df['state'].isin(north_states)
]['pollutant_avg']

south = df[
    df['state'].isin(south_states)
]['pollutant_avg']

t_stat, p_value = ttest_ind(north, south)

print("T-Statistic:", t_stat)
print("P-Value:", p_value)

if p_value < 0.05:
    print("\nConclusion:")
    print("Pollution levels in North India and South India are significantly different.")
else:
    print("\nConclusion:")
    print("No significant difference exists between North and South India pollution levels.")
```

```python
# HYPOTHESIS TEST 3
# Pollution Category vs Pollutant Type
# Chi-Square Test
print("HYPOTHESIS TEST 3")
print("Pollution Category vs Pollutant Type")
contingency_table = pd.crosstab(
    df['pollution_category'],
    df['pollutant_id']
)

chi2, p_value, dof, expected = chi2_contingency(
    contingency_table
)

print("Chi-Square Statistic:", chi2)
print("P-Value:", p_value)
print("Degrees of Freedom:", dof)

if p_value < 0.05:
    print("\nConclusion:")
    print("Pollution category and pollutant type are significantly associated.")
else:
    print("\nConclusion:")
    print("No significant association exists between pollution category and pollutant type.")
```

```python
# CORRELATION ANALYSIS
print("CORRELATION ANALYSIS")
correlation_matrix = df[
    [
        'pollutant_min',
        'pollutant_max',
        'pollutant_avg',
        'pollution_range'
    ]
].corr()

print(correlation_matrix)
```

```python
# PEARSON CORRELATION TEST
# pollutant_avg vs pollution_range
print("PEARSON CORRELATION TEST")
corr_coef, p_value = pearsonr(
    df['pollutant_avg'],
    df['pollution_range']
)

print("Correlation Coefficient:", corr_coef)
print("P-Value:", p_value)

if p_value < 0.05:
    print("\nConclusion:")
    print("A statistically significant relationship exists between average pollution and pollution range.")
else:
    print("\nConclusion:")
    print("No statistically significant relationship exists between average pollution and pollution range.")

```

```python
# CORRELATION INTERPRETATION
print("CORRELATION INTERPRETATION")
if corr_coef > 0.7:
    interpretation = "Strong Positive Correlation"

elif corr_coef > 0.3:
    interpretation = "Moderate Positive Correlation"

elif corr_coef > 0:
    interpretation = "Weak Positive Correlation"

elif corr_coef < -0.7:
    interpretation = "Strong Negative Correlation"

elif corr_coef < -0.3:
    interpretation = "Moderate Negative Correlation"

else:
    interpretation = "Weak or No Correlation"

print("Interpretation:", interpretation)
```
